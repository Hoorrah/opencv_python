```python
#using the isbnlib ( Google Books servie API _ Wikipedia API _ OpenLibrary API)
from isbnlib import *
import json

inputtype = input("Write tyepe of you are searching for book: ISBN --> i / Title -->t : ")


try:
    if (inputtype == "i"):
        get_isbn =  input("Write the ISBN:")
        googlebook = meta(get_isbn, service='goob')
        wikipedia = meta(get_isbn, service='wiki')
        openlibrary = meta(get_isbn, service='openl')
        print("\n Using Google Books service\n")
        print (googlebook)
        print("\n Using Wikipedia \n")
        print(wikipedia)
        print("\n Using OpenLibrary.org \n") 
        print(openlibrary)
        print("\n Description: ")
        print(desc(get_isbn))
        print("\n Link for cover of image:")
        print(cover(get_isbn))


    elif (inputtype == "t"):
        inputdata =  input("Write Title of the book:")
        get_isbn = isbn_from_words(inputdata)
        googlebook = meta(get_isbn, service='goob')
        wikipedia = meta(get_isbn, service='wiki')
        openlibrary = meta(get_isbn, service='openl')
        print("\n Using Google Books service\n")
        print (googlebook)
        print("\n Using Wikipedia \n")
        print(wikipedia)
        print("\n Using OpenLibrary.org \n") 
        print(openlibrary)
        print("\n Description: ")
        print(desc(get_isbn))
        print("\n Link for cover of image:")
        print(cover(get_isbn))

    else:
        print("invalid input")
        
except:
    print("There is no result")
```

```python
# using google book API
import json
from urllib.request import urlopen

def remove(string):
    return string.replace(" ", "")

inputtype = input("Write tyepe of you are searching for book: ISBN --> i / Title -->t : ")


try:
    if (inputtype == "i"):
        # create getting started variables
        api = "https://www.googleapis.com/books/v1/volumes?q=isbn:"
        isbn =  input("Enter 10 digit ISBN: ").strip()
        # send a request and get a JSON response
        resp = urlopen(api + str(remove(isbn)))
        # parse JSON into Python as a dictionary
        book_data = json.load(resp)
        volume_info = book_data["items"][0]["volumeInfo"] 
            
    elif (inputtype == "t"):
        # create getting started variables
        api = "https://www.googleapis.com/books/v1/volumes?q=title:"
        isbn =  input("Enter Title of the book: ").strip()
        # send a request and get a JSON response
        resp = urlopen(api + str(remove(isbn)))
        # parse JSON into Python as a dictionary
        book_data = json.load(resp)
        volume_info = book_data["items"][0]["volumeInfo"] 
    else:
        print("invalid Type for search")

    try:
        print(f"Title: {volume_info['title']}")
    except: pass
    try:
        print(f"Language: {volume_info['language']}")
    except: pass
    try:
        author = volume_info["authors"]
        # practice with conditional expressions!
        prettify_author = author if len(author) > 1 else author[0]
        #display title, author, page count, publication date
        print(f"Author: {prettify_author}")
    except: pass
    try:
        print(f"Publisher: {volume_info['publisher']}")
    except: pass
    try:
        print(f"ISBN_13: {volume_info['industryIdentifiers'][0]['identifier']}")
    except: pass
    try:
        print(f"ISBN_10: {volume_info['industryIdentifiers'][1]['identifier']}")
    except: pass
    try:
        print(f"Page Count: {volume_info['pageCount']}")
    except: pass
    try:
        print(f"Publication Date: {volume_info['publishedDate']}")
    except: pass
    try:
        print(f"PreviewLink: {volume_info['previewLink']}")
    except: pass
    try:
        print(f"Description: {volume_info['description']}")
    except: pass

except:
    print("It can't find that book")
     
```
